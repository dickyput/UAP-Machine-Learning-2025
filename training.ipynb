```python
# %% [markdown]
# # UAP Klasifikasi Kualitas Air (Water Quality)
# Nama: [Isi Nama Anda]
# NIM: [Isi NIM Anda]
#
# Notebook ini melatih 3 Model sesuai ketentuan modul:
# 1. CNN Scratch (Non-Pretrained)
# 2. MobileNetV2 (Transfer Learning 1)
# 3. ResNet50V2 (Transfer Learning 2)

# %%
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt
import numpy as np
import os
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns

# Pastikan folder ini sesuai hasil script susun_dataset.py
DATA_DIR = 'Dataset_Final' 
IMG_SIZE = (224, 224)
BATCH_SIZE = 32

# %% [markdown]
# ## 1. Persiapan Data (Augmentasi & Split)

# %%
# Augmentasi Data & Split Otomatis (80% Train, 20% Val)
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=30,
    width_shift_range=0.2,
    height_shift_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest',
    validation_split=0.2 
)

print("=== DATA TRAINING ===")
train_generator = train_datagen.flow_from_directory(
    DATA_DIR,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    subset='training',
    shuffle=True
)

print("\n=== DATA VALIDASI ===")
val_generator = train_datagen.flow_from_directory(
    DATA_DIR,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    subset='validation',
    shuffle=False
)

# Simpan nama kelas
class_names = list(train_generator.class_indices.keys())
print(f"\nKELAS: {class_names}")

# %% [markdown]
# ## 2. Pembangunan 3 Model

# %%
# --- MODEL 1: CNN SCRATCH ---
def build_model_scratch():
    model = tf.keras.models.Sequential([
        tf.keras.layers.Conv2D(32, (3,3), activation='relu', input_shape=(224, 224, 3)),
        tf.keras.layers.MaxPooling2D(2,2),
        tf.keras.layers.Conv2D(64, (3,3), activation='relu'),
        tf.keras.layers.MaxPooling2D(2,2),
        tf.keras.layers.Conv2D(128, (3,3), activation='relu'),
        tf.keras.layers.MaxPooling2D(2,2),
        tf.keras.layers.Flatten(),
        tf.keras.layers.Dense(512, activation='relu'),
        tf.keras.layers.Dropout(0.5),
        tf.keras.layers.Dense(len(class_names), activation='softmax')
    ])
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    return model

# --- MODEL 2: MOBILENETV2 ---
def build_model_mobilenet():
    base = tf.keras.applications.MobileNetV2(input_shape=(224, 224, 3), include_top=False, weights='imagenet')
    base.trainable = False 
    
    model = tf.keras.models.Sequential([
        base,
        tf.keras.layers.GlobalAveragePooling2D(),
        tf.keras.layers.Dense(128, activation='relu'),
        tf.keras.layers.Dropout(0.2),
        tf.keras.layers.Dense(len(class_names), activation='softmax')
    ])
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    return model

# --- MODEL 3: RESNET50V2 ---
def build_model_resnet():
    base = tf.keras.applications.ResNet50V2(input_shape=(224, 224, 3), include_top=False, weights='imagenet')
    base.trainable = False
    
    model = tf.keras.models.Sequential([
        base,
        tf.keras.layers.GlobalAveragePooling2D(),
        tf.keras.layers.Dense(128, activation='relu'),
        tf.keras.layers.Dropout(0.2),
        tf.keras.layers.Dense(len(class_names), activation='softmax')
    ])
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    return model

# %% [markdown]
# ## 3. Eksekusi Training

# %%
if not os.path.exists('models'): os.makedirs('models')

print("Training Model 1 (Scratch)...")
model_1 = build_model_scratch()
hist_1 = model_1.fit(train_generator, validation_data=val_generator, epochs=7)
model_1.save('models/model_scratch.h5')

print("\nTraining Model 2 (MobileNetV2)...")
model_2 = build_model_mobilenet()
hist_2 = model_2.fit(train_generator, validation_data=val_generator, epochs=5)
model_2.save('models/model_mobilenet.h5')

print("\nTraining Model 3 (ResNet50V2)...")
model_3 = build_model_resnet()
hist_3 = model_3.fit(train_generator, validation_data=val_generator, epochs=5)
model_3.save('models/model_resnet.h5')

# %% [markdown]
# ## 4. Evaluasi & Analisis

# %%
def evaluasi_lengkap(model, history, nama_model):
    print(f"\n--- ANALISIS: {nama_model} ---")
    
    # Grafik
    acc = history.history['accuracy']
    val_acc = history.history['val_accuracy']
    loss = history.history['loss']
    val_loss = history.history['val_loss']
    
    plt.figure(figsize=(12, 4))
    plt.subplot(1, 2, 1); plt.plot(acc, label='Train'); plt.plot(val_acc, label='Val'); plt.title(f'Accuracy {nama_model}'); plt.legend()
    plt.subplot(1, 2, 2); plt.plot(loss, label='Train'); plt.plot(val_loss, label='Val'); plt.title(f'Loss {nama_model}'); plt.legend()
    plt.show()
    
    # Classification Report
    val_generator.reset()
    Y_pred = model.predict(val_generator)
    y_pred = np.argmax(Y_pred, axis=1)
    
    print(classification_report(val_generator.classes, y_pred, target_names=class_names))
    
    # Confusion Matrix
    cm = confusion_matrix(val_generator.classes, y_pred)
    plt.figure(figsize=(5,5))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=class_names, yticklabels=class_names)
    plt.show()

# Jalankan Evaluasi
evaluasi_lengkap(model_1, hist_1, "Custom CNN")
evaluasi_lengkap(model_2, hist_2, "MobileNetV2")
evaluasi_lengkap(model_3, hist_3, "ResNet50V2")
```
